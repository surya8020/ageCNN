```python
import matplotlib.pyplot as plt # plotting
import numpy as np # linear algebra
import os # accessing directory structure
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
```

## EDA

```python
df = pd.read_csv("archive_age_gender.csv")
```

```python
np.fromstring(df.iloc[0]["pixels"],dtype=int, sep=' ').shape
```

```python
sns.countplot(x='age',data = df)
```

```python
X=df['pixels'].apply(lambda x:  np.array(x.split(), dtype="float32")) #converting data to numpy array
X=np.array(X) #normalization
print(X[0].shape)
X_t = []
for i in range(X.shape[0]):
    X_t.append(X[i].reshape(48,48,1)) #reshaping the data to (n,48,48,1)
X = np.array(X_t)

age=df['age'].values
```

```python
def plot(X,y):
    for i in range(20):
        plt.title(y[i],)
        plt.imshow(X[i].reshape(48,48))
        plt.show()
```

```python
plot(X,age)
```
